```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
import scipy.signal as signal


def plot_step_response(zeta=0.5, omega_n=1.0, K=5):

    a = 2 * zeta * omega_n
    b = omega_n ** 2
    num = [K]
    den = [1, a, b]

    system = signal.TransferFunction(num, den)
    t, y = signal.step(system)

    plt.figure(figsize=(10, 5))
    plt.plot(t, y, label=f"ζ={zeta}, ωₙ={omega_n}, K={K}")
    plt.xlabel("Time")
    plt.ylabel("Output")
    plt.grid(True)
    plt.legend()
    plt.show()

interact(
    plot_step_response,
    zeta=widgets.FloatSlider(min=0.1, max=2.0, step=0.1, value=0.7, description='ζ'),
    omega_n=widgets.FloatSlider(min=0.1, max=10.0, step=0.1, value=2.0, description='ωₙ'),
    K=widgets.FloatSlider(min=1, max=20, step=1, value=10, description='K')
);
```
